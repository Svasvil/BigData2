```python

# Lab de numeros de loteria.
import numpy as np
numeros_loteria = np.array([
    [5, 12, 22, 33, 42, 49],
    [3, 15, 22, 28, 33, 45],
    [7, 12, 19, 25, 33, 40],
    [5, 18, 22, 28, 37, 42],
    [9, 12, 22, 33, 42, 48],
    [5, 15, 20, 28, 33, 45],
    [6, 12, 22, 34, 40, 49],
    [5, 17, 21, 28, 33, 42],
    [8, 12, 22, 29, 33, 47],
    [5, 14, 22, 30, 33, 41]
])


def contar_Frecuencia(numeros_loteria):
    frecuencia = {}
    for fila in numeros_loteria:
        for numero in fila:
            if numero in frecuencia:
                frecuencia[numero] += 1
            else:
                frecuencia[numero] = 1
    return frecuencia

def numero_mas_Frecuente(frecuencia,top_n=5):
    frecuencia_ordenada = sorted(frecuencia.items(), key=lambda x: x[1], reverse=True)
    return frecuencia_ordenada

def generar_numeros_sugeridos(numeros_frecuentes, total_numeros=6):
    sugeridos = [numero[0] for numero in numeros_frecuentes]
    while len(sugeridos) < total_numeros:
        numero_aleatorio = np.random.randint(1, 50)
        if numero_aleatorio not in sugeridos:
            sugeridos.append(numero_aleatorio)
    return sugeridos

frecuencias1 = contar_Frecuencia(numeros_loteria)
Top_numeros = numero_mas_Frecuente(frecuencias1,top_n=5)
numeros_sugeridos = generar_numeros_sugeridos(Top_numeros)
print(numeros_sugeridos)

print(" \n Frecuencias de los números en la lotería:", frecuencias1)
print(" \n Los números más frecuentes son:", Top_numeros)
print("\n Números sugeridos para apostar:", numeros_sugeridos)
```

```python
#Practica de num de listas
import numpy as np
listaNums = np.array ([23,53,67,19,12,5,1,16,50,100])

def sumaNumeros(listaNums):
    suma = 0
    for numero in listaNums:
        suma += numero
    return suma
print("la suma de los numeros es ",sumaNumeros(listaNums))

def numsPares(listaNums):
    pares = []
    for numero in listaNums:
        if numero % 2 == 0:
            pares.append(numero)
    return pares
print("los numeros pares son ",numsPares(listaNums))

def numMayor(listaNums):
    cant = 0
    for i in listaNums:
        if i > 20:
            cant += 1
    return cant
print("la cantidad de numeros mayores a 20 es ", numMayor(listaNums))
```

```python
#Aca lo que estamos haciendo es basicamten reconvertir un tipo de dato , si vemos acabamos de reconvertir el Genero de masculino y femenino , a un 1 y 0
for col in DataFrame.select_dtypes(include=['object']).columns:
  le=LabelEncoder()
  DataFrame[col]=le.fit_transform(DataFrame[col])
  display(DataFrame.head())
```

```python
#hacemos que los datos queden cerca del cero , entre mas cercano a cero seria un caso o el otro , por ejemplo los true podemos ver que son mujeres
numeric_columns = DataFrame.select_dtypes(include=['number']).columns
scaler = StandardScaler()
DataFrame[numeric_columns] = DataFrame[numeric_columns]-DataFrame[numeric_columns].mean()
display(DataFrame.head())
```

```python
#lo que hacemos es buscar los valores que esten dentro del rango para ver cuales funcionan
scaler = StandardScaler()
DataFrame[numeric_columns] = scaler.fit_transform(DataFrame[numeric_columns])
display(DataFrame.head())
```

```python
#Eliminamos los datos atipicos , esto para tener los datos de tal rango , en este los datos que son menores al 0.25 y mayores al 0.75 seran atipicos.
Quartil1 = DataFrame[numeric_columns].quantile(0.25)
Quartil3 = DataFrame[numeric_columns].quantile(0.75)
IQR = Quartil3 - Quartil1
Datalimpia = DataFrame[~((DataFrame[numeric_columns]< (Quartil1-1.5 * IQR)) | (DataFrame[numeric_columns] > (Quartil3 + 1.5 * IQR))).any(axis=1)]
display(Datalimpia)
```
