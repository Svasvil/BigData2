```python

# Lab de numeros de loteria.
import numpy as np
numeros_loteria = np.array([
    [5, 12, 22, 33, 42, 49],
    [3, 15, 22, 28, 33, 45],
    [7, 12, 19, 25, 33, 40],
    [5, 18, 22, 28, 37, 42],
    [9, 12, 22, 33, 42, 48],
    [5, 15, 20, 28, 33, 45],
    [6, 12, 22, 34, 40, 49],
    [5, 17, 21, 28, 33, 42],
    [8, 12, 22, 29, 33, 47],
    [5, 14, 22, 30, 33, 41]
])


def contar_Frecuencia(numeros_loteria):
    frecuencia = {}
    for fila in numeros_loteria:
        for numero in fila:
            if numero in frecuencia:
                frecuencia[numero] += 1
            else:
                frecuencia[numero] = 1
    return frecuencia

def numero_mas_Frecuente(frecuencia,top_n=5):
    frecuencia_ordenada = sorted(frecuencia.items(), key=lambda x: x[1], reverse=True)
    return frecuencia_ordenada

def generar_numeros_sugeridos(numeros_frecuentes, total_numeros=6):
    sugeridos = [numero[0] for numero in numeros_frecuentes]
    while len(sugeridos) < total_numeros:
        numero_aleatorio = np.random.randint(1, 50)
        if numero_aleatorio not in sugeridos:
            sugeridos.append(numero_aleatorio)
    return sugeridos

frecuencias1 = contar_Frecuencia(numeros_loteria)
Top_numeros = numero_mas_Frecuente(frecuencias1,top_n=5)
numeros_sugeridos = generar_numeros_sugeridos(Top_numeros)
print(numeros_sugeridos)

print(" \n Frecuencias de los números en la lotería:", frecuencias1)
print(" \n Los números más frecuentes son:", Top_numeros)
print("\n Números sugeridos para apostar:", numeros_sugeridos)
```

```python
#Practica de num de listas
import numpy as np
listaNums = np.array ([23,53,67,19,12,5,1,16,50,100])

def sumaNumeros(listaNums):
    suma = 0
    for numero in listaNums:
        suma += numero
    return suma
print("la suma de los numeros es ",sumaNumeros(listaNums))

def numsPares(listaNums):
    pares = []
    for numero in listaNums:
        if numero % 2 == 0:
            pares.append(numero)
    return pares
print("los numeros pares son ",numsPares(listaNums))

def numMayor(listaNums):
    cant = 0
    for i in listaNums:
        if i > 20:
            cant += 1
    return cant
print("la cantidad de numeros mayores a 20 es ", numMayor(listaNums))
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
filepath="/content/calificaciones.csv"
data = pd.read_csv(filepath)
print (data.head())

#Normalizamos los datos.
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
#aplicamos el model o de kmeans
k_optimal = 3
kmeans = KMeans(n_clusters=k_optimal, random_state=42)
kmeans.fit(data_scaled)
data["cluster"] = kmeans.labels_
#vemos los datos
plt.figure(figsize=(14,7))
sns.scatterplot(
    x='Calificación en Literatura',
    y='Calificación en Álgebra',
    hue='cluster',
    data=data,
)
plt.title("Clusters de Estudiantes")
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://sistemas.ufidelitas.ac.cr/tcu/'

headers = {
    'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36'
}

try:
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    print(response)
    soup = BeautifulSoup(response.text, 'html.parser')

    preguntas = []
    respuestas = []

    toggle_items = soup.find_all('div', class_='elementor-toggle-item')

    for item in toggle_items:
        pregunta = item.find('a', class_='elementor-toggle-title').get_text(strip=True)
        respuesta = item.find('div', class_='elementor-tab-content').get_text(separator=' ', strip=True)

        preguntas.append(pregunta)
        respuestas.append(respuesta)

    df = pd.DataFrame({'Pregunta': preguntas, 'Respuesta': respuestas})
    print(df)
```

```python
#Semana 11
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
file_path =""
data=pd.read_csv("/content/desempeno_estudiantes.csv")
print(data.head())


x= data[['Student_ID' ,'Attendance_Rate','Average_Grade','Late_Submissions','Extracurricular_Activities']]

# Separate features to be scaled from the Student_ID
x_features = data[['Attendance_Rate','Average_Grade','Late_Submissions','Extracurricular_Activities']]

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x_features)

inercia = []
max_clusters = len(x_features) # Número máximo de clusters igual al número de muestras
for k in range(1, max_clusters + 1): # Ajustar el rango del bucle
    kmeans_temp = KMeans(n_clusters=k, random_state=42) # Crear un modelo temporal
    kmeans_temp.fit(x_scaled)
    inercia.append(kmeans_temp.inertia_)

# Graficar el método del codo
plt.plot(range(1, max_clusters + 1), inercia, marker='o')
plt.title("Método del Codo")
plt.xlabel("Número de Clusters")
plt.ylabel("Inercia")
plt.show()

optimal_k = 2
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
data["cluster"] = kmeans.fit_predict(x_scaled)
print(data)

pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)
# Graficar los clusters
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=data["cluster"], cmap='viridis')
plt.title("Segmentación de Estudiantes")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.colorbar(label="Cluster")
plt.show()
```
